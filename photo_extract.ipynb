```python
%pip install opencv-python
import cv2
import os

# Path to the video file directory (or a direct video file path)
#video_dir = r"D:\onvif-021\20250626AM"
video_dir = r"C:\Users\pfh3221\OneDrive - Université de Moncton\AI_monkey\Generic_ONVIF-009-20241102-165954-1730581194385-7.mp4"

video_extensions = ('.mp4', '.avi', '.mov', '.mkv')
video_path = None

# If video_dir is already a file, use it directly
if os.path.isfile(video_dir) and video_dir.lower().endswith(video_extensions):
    video_path = video_dir
# If video_dir is a directory, pick the first video file inside it
elif os.path.isdir(video_dir):
    video_files = [f for f in os.listdir(video_dir) if f.lower().endswith(video_extensions)]
    if video_files:
        video_path = os.path.join(video_dir, video_files[0])
    else:
        print(f"No video files found in {video_dir}")
else:
    print(f"Invalid path: {video_dir}")

if video_path:
    print(f"Opening video: {video_path}")
    
    # Open the video file
    cap = cv2.VideoCapture(video_path)
    
    if cap.isOpened():
        print("Video opened successfully!")
        print(f"Total frames: {int(cap.get(cv2.CAP_PROP_FRAME_COUNT))}")
        print(f"FPS: {cap.get(cv2.CAP_PROP_FPS)}")
        print(f"Resolution: {int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))}x{int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))}")
    else:
        print("Error: Could not open video file")
```

```python
import time
from datetime import datetime
import numpy as np

# Output directory for extracted frames
output_dir = r"C:\Users\pfh3221\OneDrive - Université de Moncton\AI_monkey"

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Target resolution
target_width = 3840
target_height = 2160

# Extract naming portion from video filename
video_filename = os.path.basename(video_path)
# Find "ONVIF-" in the filename and extract 19 characters after it
onvif_index = video_filename.find("ONVIF-")
if onvif_index != -1:
    # Extract 19 characters starting after "ONVIF-"
    name_part = video_filename[onvif_index + 6:onvif_index + 6 + 19]
else:
    # Fallback if "ONVIF-" not found
    name_part = "unknown"

# Reopen the video to ensure we start from the beginning
cap = cv2.VideoCapture(video_path)

print(f"Output directory: {output_dir}")
print(f"Target resolution: {target_width}x{target_height}")
print(f"File naming pattern: Frame1-{name_part}")

# Read the first frame
ret, frame = cap.read()

if ret:
    print(f"Frame read successfully. Original size: {frame.shape[1]}x{frame.shape[0]}")

    # Resize frame to target resolution
    resized_frame = cv2.resize(frame, (target_width, target_height))
    print(f"Resized to: {resized_frame.shape[1]}x{resized_frame.shape[0]}")

    # Generate filename using video name pattern
    filename = f"Frame1-{name_part}.jpg"
    filepath = os.path.join(output_dir, filename)

    # Use imencode + Python file write to support Unicode paths (cv2.imwrite fails on accented chars)
    success, encoded_img = cv2.imencode('.jpg', resized_frame)
    if success:
        with open(filepath, 'wb') as f:
            f.write(encoded_img.tobytes())
        file_size = os.path.getsize(filepath)
        print(f"Extracted: {filename} ({file_size / 1024:.1f} KB) -> {filepath}")
    else:
        print(f"ERROR: cv2.imencode() failed to encode the image.")
else:
    print("Error: Could not read frame from video")

# Release the video capture object
cap.release()
print(f"\nExtraction complete! 1 frame extracted." if ret and success else "\nExtraction failed.")
```
