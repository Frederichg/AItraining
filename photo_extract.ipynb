```python
%pip install opencv-python
import cv2
import os

# Path to the video file directory
video_dir = r"D:\onvif-021\20250626AM"

# Get the first video file in the directory
video_files = [f for f in os.listdir(video_dir) if f.endswith(('.mp4', '.avi', '.mov', '.mkv'))]

if video_files:
    video_path = os.path.join(video_dir, video_files[0])
    print(f"Opening video: {video_path}")
    
    # Open the video file
    cap = cv2.VideoCapture(video_path)
    
    if cap.isOpened():
        print(f"Video opened successfully!")
        print(f"Total frames: {int(cap.get(cv2.CAP_PROP_FRAME_COUNT))}")
        print(f"FPS: {cap.get(cv2.CAP_PROP_FPS)}")
        print(f"Resolution: {int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))}x{int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))}")
    else:
        print("Error: Could not open video file")
else:
    print(f"No video files found in {video_dir}")
```

```python
import time
from datetime import datetime

# Output directory for extracted frames
output_dir = r"D:\Robotform_photo"

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Target resolution
target_width = 3840
target_height = 2160

# Extract naming portion from video filename
video_filename = os.path.basename(video_path)
# Find "ONVIF-" in the filename and extract 19 characters after it
onvif_index = video_filename.find("ONVIF-")
if onvif_index != -1:
    # Extract 19 characters starting after "ONVIF-"
    name_part = video_filename[onvif_index + 6:onvif_index + 6 + 19]
else:
    # Fallback if "ONVIF-" not found
    name_part = "unknown"

# Reopen the video to ensure we start from the beginning
cap = cv2.VideoCapture(video_path)

# Get video FPS to calculate frame interval
fps = cap.get(cv2.CAP_PROP_FPS)
# If FPS is 0 or invalid, use default of 25 fps
if fps == 0 or fps is None:
    fps = 25.0
    print(f"Warning: Video FPS not available, using default {fps} fps")
    
frame_interval = int(fps * 60)  # Extract every 60 seconds

print(f"Extracting frames every 60 seconds (every {frame_interval} frames at {fps} fps)")
print(f"Output directory: {output_dir}")
print(f"Target resolution: {target_width}x{target_height}")
print(f"File naming pattern: Frame#-{name_part}")

frame_count = 0
extracted_count = 1  # Start from 1 for frame numbering

while cap.isOpened():
    ret, frame = cap.read()
    
    if not ret:
        break
    
    # Extract frame every 60 seconds
    if frame_count % frame_interval == 0:
        # Resize frame to target resolution
        resized_frame = cv2.resize(frame, (target_width, target_height))
        
        # Generate filename using video name pattern
        filename = f"Frame{extracted_count}-{name_part}.jpg"
        filepath = os.path.join(output_dir, filename)
        
        # Save the frame as JPG
        cv2.imwrite(filepath, resized_frame)
        print(f"Extracted: {filename}")
        extracted_count += 1
    
    frame_count += 1

# Release the video capture object
cap.release()
print(f"\nExtraction complete! Total frames extracted: {extracted_count - 1}")
```
